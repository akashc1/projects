### Task 1. Image Denoising and edge detection
In this task I used a Gaussian filter and its derivative in order to denoise and find edges in an
image. I varied the standard deviation of the Gaussian filter in order to see the trade-off
between denoising and the preservation of high-frequency content.

* _Denoising_
Loading the first image of the sequence and create a two-dimensional, separable and circularly symmetric Gaussian filter of the form:
\begin{equation}
	g_{mn} = g_m g_n,\quad g_n = e^{-\frac{1}{2}(\frac{n}{\sigma})^2},\quad m,n=-20,\ldots,20,
\end{equation}
where $\sigma$ controls the width of the filter. Filtering the image with $g_{mn}$ for values of $\sigma = 1,2,4$.
Displaying the original and filtered images and plot the shape of the 2D filter.

##### Observations:
  * A value of $\sigma = 4$ is better in reducing high-frequency noise. A filter with a value of $\sigma = 1$ is most narrow in the frequency domain.
  * Increasing $\sigma$ obviously eliminates the most noise, but you definitely lose the density of features one can extract (edges, etc.)

```python

x = np.arange(-20, 21)
y = np.arange(-20, 21)

plt.imshow(imglist[0])
plt.title('Original MRI image')


# iterate through exponents for sigma values
for i in range(0, 3):
    sig = 2 ** i
    g_n = []
    g_m = []
    
    # create gaussian filters
    for i in x:
        g_n.append(np.exp((-0.5) * ((i / sig) ** 2)))
        g_m.append(np.exp((-0.5) * ((i / sig) ** 2)))
                   
    g_n = np.transpose(g_n)
                   
    g_mn = np.outer(g_m, g_n)
                   
    filtered_image = signal.convolve(imglist[0], g_mn, mode='same')
    
    plt.figure()
    plt.imshow(filtered_image)
    plt.title('Image with Gaussian Filter of sigma = ' + str(sig))
```

* _Edge Detection_
Detecting edges in the image using a derivative of Gaussian filter that smoothes the image in one direction while detecting the edges in the other. For instance, to detect edges in the horizontal dimension I use the filter:
$$
 \gamma_{mn} = g_m d_n,
$$
where $d_n = -\frac{2}{\sigma^2} n e^{-\frac{1}{2}(\frac{n}{\sigma})^2}$ for $n=-6,\ldots,6$. For the smoothing part $g_n$ I use $\sigma = 0.75$ while for the differentiation part $d_n$ I use $\sigma = 2$. Filtering the image, normalizing and thresholding it, and using non-maximum suppression over the horizontal dimension in order to detect the edges. Plotting the images at all intermediate steps, and repeating the procedure for the vertical dimension. Then I combine them into a single one by just using a logical OR

```python

index_arr = [i for i in range(-6, 7)]

def derv(n, sigma):
    return ((-2 / sigma**2) * n * np.exp(-0.5 * ((n**2) / (sigma**2))))


g_sigma = 0.75
d_sigma = 2


d_n = []
d_m = []

g_n = []
g_m = []
    
for i in range(-6, 7):
    g_n.append(np.exp((-0.5) * ((i / g_sigma) ** 2)))
    g_m.append(np.exp((-0.5) * ((i / g_sigma) ** 2)))

# create derivative filters
for i in range (-6, 7):
    d_n.append(derv(i, d_sigma))
    d_m.append(derv(i, d_sigma))

d_n = np.transpose(d_n)

g_m = np.transpose(g_m)

# create combined filters
h_edge_filter = np.outer(g_m, d_n)
v_edge_filter = np.outer(d_m, g_n)

num_cols = imglist[0].shape[1]
num_rows = imglist[0].shape[0]


# calculate horizontal edges
h_edges = signal.convolve2d(imglist[0], h_edge_filter, mode='same')
plt.figure()
plt.imshow(h_edges)
plt.title('Horizontal Edge Filter')

h_mu = np.mean(h_edges)
h_sig = np.std(h_edges)
norm_h_edges = (h_edges - h_mu) / h_sig

plt.figure()
plt.imshow(norm_h_edges)
plt.title('Normalized Horizontal Edge Filter')

abs_h_edges = np.absolute(norm_h_edges)

# thresholding using the standard deviation
thres_h_edges = norm_h_edges
for i in range (len(thres_h_edges)):
    for j in range (len(thres_h_edges)):
        if norm_h_edges[i,j] < 1.5*np.std(norm_h_edges):
            thres_h_edges[i,j] = 0
            

            
plt.figure()
plt.imshow(thres_h_edges)
plt.title('Thresholded Horizontal Edge Filter')

num_cols = abs_h_edges.shape[1]
num_rows = abs_h_edges.shape[0]

# non-max suppression
sup_h_edges = np.zeros((num_cols, num_rows))
for i in range(0, num_cols):
    for j in range(0, num_rows):
        if (thres_h_edges[i][j] > thres_h_edges[i][j - 1] and thres_h_edges[i][j] > thres_h_edges[i][j + 1]):
            sup_h_edges[i][j] = 1


plt.figure()
plt.imshow(sup_h_edges)
plt.title('Max-Suppressed Horizontal Edge Filter')      


plt.figure()
# finding vertical edges
v_edges = signal.convolve2d(imglist[0], v_edge_filter, mode='same')
plt.imshow(v_edges)
plt.title('Vertical Edge Filter')

v_mu = np.mean(v_edges)
v_sig = np.std(v_edges)
norm_v_edges = (v_edges - v_mu) / v_sig

plt.figure()
plt.imshow(norm_v_edges)
plt.title('Normalized Vertical Edge Filter')

abs_v_edges = np.absolute(norm_v_edges)

# thresholding
thres_v_edges = norm_v_edges
for i in range (len(thres_v_edges)):
    for j in range (len(thres_v_edges)):
        if norm_v_edges[i,j] < 1.5*np.std(norm_v_edges):
            thres_v_edges[i,j] = 0
            
plt.figure()
plt.imshow(thres_v_edges)
plt.title('Thresholded Vertical Edge Filter')

# non-max suppression
sup_v_edges = np.zeros((num_cols, num_rows))
for i in range(0, num_cols):
    for j in range(0, num_rows):
        if thres_v_edges[i][j] > thres_v_edges[i - 1][j] and thres_v_edges[i][j] > thres_v_edges[i + 1][j]:
            sup_v_edges[i][j] = 1


plt.figure()
plt.imshow(sup_v_edges)
plt.title('Max-Suppressed Vertical Edge Filter') 

# combining vertical and horizontal edges
all_edges = np.logical_or(sup_v_edges,sup_h_edges)
plt.figure(figsize=(10,10))
plt.imshow(all_edges)
plt.title('Combined Edge Detection') 
```

* _Histogram representation_
Building a histogram of gradients at the detected edge points. The magnitude (strength) and angle of an edge at position $m,n$ is computed as:

   $$ s = \sqrt{ (\nabla_x I_{mn})^2 + (\nabla_y I_{mn})^2},\quad\theta = \arctan\big(\nabla_y I_{mn}/\nabla_x I_{mn}\big),$$
   
   where $\nabla_x$ and $\nabla_y$ represent the derivatives of the signal (e.g. image $I$) over the horizontal and vertical dimensions, respectively. Building two histograms, one over the intensities and another over the angles. Using $40$ bins for building the histograms and plotting the resulting representations.
   * The histogram of angles makes reasonable sense since the edges in the previous figures seem to cover most angles, and the histogram shows a relatively even distribution.


```python
# find all edges as binary
uniform_edges = (all_edges > 0)

plt.figure()
plt.imshow(uniform_edges)
plt.title('All Edges')

x_grad = h_edges[uniform_edges]
y_grad = v_edges[uniform_edges]

# calculate intensity
mag = np.sqrt((x_grad ** 2) + (y_grad ** 2))

plt.figure()
plt.hist(mag, 40)
plt.title('Histogram of Intensities')

# calculate angles
theta = np.arctan(y_grad / x_grad)
plt.figure()
plt.hist(theta, 40)
plt.title('Histogram of Angles')
```

### Task 2. Data reduction and basis representation
* _Columns as data points_
Using again the first image of the sequence and considering its representation in the Fourier and PCA
basis over the horizontal axis, which means considering each column of the image as a different data point:

   $$I = \big[\boldsymbol x_1,\boldsymbol x_2,\ldots,\boldsymbol x_c\big],$$
    
where $\boldsymbol x_i\in\mathbb{C}^D$ is the $i$th column of the image and $c$ is the number of columns.
Let $\widehat{I}^{(d)}$ be the reconstruction of the image using the first $d$ basis elements and define the reconstruction error as $e_d = \lVert I - \widehat{I}^{(d)} \rVert^2$. Recall that if $\boldsymbol U$ is an orthonormal basis then the expansion coefficients of a vector $\boldsymbol x$ over the basis $\boldsymbol U$ can be computed as:

   $$\boldsymbol y = \boldsymbol U^* \boldsymbol x,$$
   
   where $\boldsymbol U^*$ denotes transpose and complex conjugate.
   
The reconstruction $\widehat{\boldsymbol x}$ of $\boldsymbol x$ over $\boldsymbol U$ is computed as:

   $$\hat{\boldsymbol x} = \boldsymbol U\boldsymbol y = \boldsymbol U \boldsymbol U^* \boldsymbol x.$$

If vector $\boldsymbol x\in\operatorname{span}\big\{\boldsymbol U\big\}$ then $\widehat{\boldsymbol x} = \boldsymbol x$. Otherwise, $\widehat{\boldsymbol x}$ is the orthogonal projection of $\boldsymbol x$ onto the range space of $\boldsymbol U$ (i.e. $\widehat{\boldsymbol x} = \arg\min_{\boldsymbol z}\lVert\boldsymbol x- \boldsymbol z\rVert$).

Computing the covariance matrix of the data points and extracting the PCA basis (eigenvectors of the covariance matrix). Calculating the reconstruction error sequence for both Fourier and PCA basis and plot it over the number of components used. For the Fourier case, considering $d=1,3,\ldots,15$ and signal reconstruction using conjugate frequency pairs.

  * The PCA basis produces the smallest reconstruction error for a given number of components $d$, since it is a data-specific basis created by the individual data points. It could be less convenient to use a PCA basis for this reason, but it's definitely better for this kind of application.

```python
image = imglist[0]

covariance = np.cov(image, rowvar='False')

# derive eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance)
e_vectors_t = np.transpose(eigenvectors)

pca_error = []

# find coefficients, reconstructed image, and corresponding error
for i in range(len(image)):
    e_coeffs = np.matmul(e_vectors_t, image)
    e_coeffs[i:][:] = 0
    img_reconstruct = np.matmul(eigenvectors, e_coeffs)
    pca_error.append((np.linalg.norm(image - img_reconstruct)) ** 2)
    
plt.figure()
plt.plot(pca_error)
plt.title('Error due to PCA Reconstruction')
plt.figure()
plt.imshow(img_reconstruct)
plt.title('PCA Reconstructed Image')


mu = np.mean(image)
sig = np.std(image)

n_image = (image - mu) / sig

# find fourier transform of the image
f_image = np.fft.fft(n_image)

f_image_reconstruct = np.zeros((np.shape(f_image)), dtype=complex)

f_error = []

# reconstruct fourier image and find error
for i in range(0, 96):
    f_image_reconstruct[:, i] = f_image[:, i]
    f_image_reconstruct[:, -i] = f_image[:, -i]
    image_reconstruct = np.fft.ifft(f_image_reconstruct)
    f_error.append((np.linalg.norm(image - image_reconstruct)) ** 2)
    
plt.figure()
plt.plot(f_error)
plt.title('Error due to Fourier Reconstruction')
plt.figure()
plt.imshow(image_reconstruct.real)
plt.title('Fourier Reconstructed Image')
```

* _Images as data points_ Repeating the above task but now using the entire image sequence and considering each image as a data point. For that purpose, stacking the vectorized images into a matrix, computing the PCA basis and calculating the reconstruction error as a function of the number of components used. Computing also the reconstruction sequence using Fourier basis. Now the data matrix is given by:

   $$\boldsymbol X = \big[\operatorname{vec}\big(I_1\big),\ldots,\operatorname{vec}\big(I_{T}\big)\big],\quad \operatorname{vec}\big(I\big) = \left[\begin{array}{c}\boldsymbol x_1\\\vdots\\\boldsymbol x_c\end{array}\right],$$
   
   where $T$ is the number of frames in the sequence.

   * The dimensionality of the data is now much larger, which results in some computation issues. For this part I'm using the ${\tt PCA}$ object of the [scikit-learn](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) library.



```python

# PCA basis
normImage = []
for i in range (15):
    normImage = imglist - np.mean(imglist, axis = 0)
# vectorize images
normImage2 = []
for i in range (15):
    vImages = (imglist[i]).flatten()
    normImage2.append(vImages)

    
# make component array
p = np.arange(1,16,2)
for i in range(len(p)):
    # find PCA components
    pca = PCA(n_components = p[i])
    # find components and reconstruct
    component = pca.fit_transform(normImage2)
    proj = pca.inverse_transform(component)
 
    # plot error
    PCAerror = (np.linalg.norm(normImage2-proj))**2
    plt.plot(p[i], PCAerror, 'ok')
    
# fourier basis
y = np.fft.fft(normImage2, axis = 0)

# reconstruct image
index = np.arange(8)
for i in range(len(p)):

    yhat = np.zeros((15,36864))
    # select the rows of the fourier basis we actually want
    yhat[np.arange(-index[i],index[i]+1),:] = y[np.arange(-index[i],index[i]+1),:]
    Xf = np.real(np.fft.ifft(yhat, axis=0))

    # calcualte and plot error
    F_error = (np.linalg.norm(normImage2-Xf))**2
    plt.plot(p[i], F_error, 'or')
    
plt.title('PCA vs Fourier Error; Black = PCA, Red = Fourier')
plt.xlabel('p value')
plt.ylabel('Error')


# Each data point is 192x192 = 36864 in size. Computational errors resulted from having non-scalar data. TO address
# this, we vectorized each image.

# The fourier basis may work better now since there is some activity based on a frequency, which is the domain of 
# fourier. This may eliminate some error and provide a better fit
```

### Task 3. Data regression and linear prediction

* _Linear prediction_
Consider the problem of predicting the current frame in the sequence by using previous frames. For this task we will be using linear regression to predict the current frame based on a linear combination of the previous two frames that is, we want to find the regression coefficients $\boldsymbol w = [w_1,w_2,w_3]$ for solving the problem:

   $$\min_{\boldsymbol w}\, \lVert \boldsymbol x^t - \widehat{\boldsymbol x}^t\rVert^2,\quad\widehat{\boldsymbol x}^t = w_1\boldsymbol x^{t-1} + w_2\boldsymbol x^{t-2} + w_3\boldsymbol 1,$$
   
   where $\boldsymbol 1$ is a constant image of all ones and where the superscript $.^t$ denotes the time-index (frame) of the sequence.
Assuming $\boldsymbol x^t = \operatorname{vec}\big(I_t\big)$ is your vectorized image, we can express the problem in the form: $\min_{\boldsymbol w}\;\lVert\boldsymbol x^t - \boldsymbol \Phi^t\boldsymbol w\rVert$ (i.e. give an expression for $\boldsymbol \Phi^t$). Finding the optimal coefficients for $t=1,\ldots,15$ and plotting the evolution of the coefficients over time. Assuming circular symmetry for estimating the coefficients for $t=1,2$ (i.e. $\widehat{\boldsymbol x}^1 = w_1\boldsymbol x^{15} + w_2\boldsymbol x^{14}+ w_3\boldsymbol 1$). Creating a plot of the prediction error at each frame.

```python
# create phi matrix
phi = np.zeros((36864,3))
phi[:,2] = 1

images = []

for i in range(15):
    images.append(imglist[i].flatten())
    
coeffs = np.zeros((15,3))

plt.figure()

# find phi matrix, (phi_t * phi)^1 and use it to get the x
# find and plot the error
for i in range(2, 15):
    phi[:, 0] = images[i - 1]
    phi[:, 1] = images[i - 2]    
    phi_phi_t = np.matmul(np.transpose(phi), phi)
    phi_phi_tinv = np.linalg.pinv(phi_phi_t)    
    coeffs[i, :] = np.matmul(np.matmul(phi_phi_tinv, np.transpose(phi)), images[i])
    err = np.linalg.norm(np.transpose(images[i]) - np.matmul(phi, coeffs[i, :]))
    plt.plot(i, err, 'or')
   
# for the edge cases when i should be 0 and 1
phi[:,0] = images[14]
phi[:,1] = images[13]
phi_phi_t = np.matmul(np.transpose(phi), phi)
phi_phi_tinv = np.linalg.pinv(phi_phi_t)    
coeffs[0, :] = np.matmul(np.matmul(phi_phi_tinv, np.transpose(phi)), images[0])
err = np.linalg.norm(np.transpose(images[0]) - np.matmul(phi, coeffs[0, :]))
plt.plot(0, err, 'or')
# edge case for 1
phi[:,0] = images[0]
phi[:,1] = images[14]
phi_phi_t = np.matmul(np.transpose(phi), phi)
phi_phi_tinv = np.linalg.pinv(phi_phi_t)    
coeffs[1, :] = np.matmul(np.matmul(phi_phi_tinv, np.transpose(phi)), images[1])
err = np.linalg.norm(np.transpose(images[1]) - np.matmul(phi, coeffs[1, :]))
plt.plot(1, err, 'or')
plt.title('Regression Error vs. Time')
plt.xlabel('Image in Sequence (time)')
plt.ylabel('Magnitude of Error')

# plot results
plt.figure()
plt.plot(coeffs[:,0],'r')
plt.plot(coeffs[:,1],'g')
plt.plot(coeffs[:,2],'b')
plt.title('Image Regression Coefficients vs. Time; red = w1, green = w2, blue = w3')
plt.xlabel('Image in Sequence (time)'); 
plt.ylabel('Weight')
```
