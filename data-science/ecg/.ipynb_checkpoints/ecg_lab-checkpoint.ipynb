### QRS complex detection
In this task we will implement a simple version of a QRS-complex detector by just setting a threshold on the
observed ECG signal. We will then use those detections to provide an estimate of the heart rate.

* **Signal Generation -** We are going to create a synthetic EGC signal that we will be using to test a peak detector algorithm. For that purpose generate a stream of $10$ equally spaced pulses over a time span of $10\ s$ and with a sampling rate of $f_s = 256\ Hz$. The goal of this task is to create a basic model for a ECG signal. You can think of the signal as the convolution of a canonical pulse shape $\varphi(t)$ with a stream of Dirac delta functions: 

   $$x(t) = \sum_{k=1}^K a_k\, \varphi(t-t_k) = \varphi(t) \ast \sum_{k=1}^K a_k\, \delta(t-t_k),$$ 

   where $K$ is the number of pulses observed, $a_k$ represent the amplitudes of the different pulses and $t_k$ correspond to the locations of the pulses in time. For the purpose of this task use the canonical pulse shape generated by the **ecg_wave($\cdot$)** function and constant unit amplitudes $a_k=1$. Make sure to scale your function in time so that its support is only $1/10\ s$. Plot the generated signal over time adding the appropriate labels for the axis.


```python
# ============================================================================
# ============================================================================
# Akash Chaurasia
# akashc@jhu.edu
# ============================================================================
# ============================================================================

# ============================================================================
# import modules
# ============================================================================
# Note that this part of the code needs to be run prior to any other code cell

import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal

# inline plots
%matplotlib inline
sns.set()

# ============================================================================
# functions
# ============================================================================
def ecg_wave(x):
    """
    This function generates a synthetic ECG template of unit length (support).
    """
    # compute signal (superposition of splines)
    return 0.3 * signal.bspline(3*3*x-7.5,2)\
         + 0.15* signal.bspline(3*4*x-2,3)\
         + signal.bspline(3*2*2*x-6,1) - 0.2 * signal.bspline(3*4*x-5,1) - 0.4 * signal.bspline(3*4*x-7,1)


# ============================================================================
# Task 1. QRS complex detection
# ============================================================================

# emulate an ECG signal as a strem of synthetic pulses
# ============================================================================

# create a sample ECG pulse from the given function over 256 samples
ecg_sample = list()
for i in range (0, 255):
    ecg_sample.append(ecg_wave(i / 256))
 

# define parameters such as sampling frequency, observation window, number of pulses
s_frequency = 256
obs_window = 10
num_pulses = 10
sample_time = np.linspace(0, 1, s_frequency - 1)

# create a time discretization
diracs = list()
for i in range (0, (s_frequency) * 10):
    diracs.append(0)
    
    # offset by 1/2 of a second to center signal
    if (i + 128) % (s_frequency) == 0:
        diracs[i] = 1

# create time list for better plotting
time = np.linspace(0, obs_window, s_frequency * num_pulses)

# plot basic pulse shape
plt.figure()
plt.plot(sample_time, ecg_sample) 
plt.title('Sample ECG Pulse')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')


# generate ECG signal by convolution the basic shape with a stream of Diracs
final_signal = np.convolve(diracs, ecg_sample, mode='same')

# display resulting waveform and add axis labels
plt.figure()
plt.plot(time, final_signal)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.title('ECG Voltage vs. Time')
```

* **Find Peaks -** Implement a QRS-complex detector by finding _"peaks"_ (local maxima or minima) or edges (sharp transitions) of the previously generated ECG signal. For that purpose you can use differentiation, thresholding and non-maximum suppression. We recommend that you normalize the signal before thresholding. For instance, you can normalize the signal to have zero mean and unit variance. After normalization define a threshold value  and keep only those values of the signal that are above the given threshold. Once you have done that, perform a non-maximum suppression (i.e. keep a value if it is greater than the previous and following values). Note that in your detection routine you want to keep one peak per QRS-complex.

```python
# you might want to implement the peak detector as a function for later use
def peak_detector(x,threshold=0.5):
    """
    This function implements a basic peak detector based on thresholding and
    differentiation.
    """

    # initializing list object
    ecg_diff = list()

    # normalization
    
    # find mean and stdev
    mu = np.mean(x)
    s = np.std(x)
    
    # append in new list the corresponding z-score
    for i in range(len(x)):
        # z-score calculation
        z = ((x[i] - mu) / s)
        ecg_diff.append(z)
    
    # thresholding
    # for every value in the list, make it zero if it is lower than the given threshold
    for i in range(len(ecg_diff)):
        if ecg_diff[i] < threshold:
            ecg_diff[i] = 0
            
    # non-maximum suppresion
    # here we look at 4 values on either side of the selected value
    # if the selected value is less than any of the 4 neighbors on either side, make it 0
    for i in range(len(ecg_diff)):
        current = ecg_diff[i]
        
        # for every value, go 4 on either side
        for j in range(-4,4):
            # this if statement keeps the indices in bounds
            if i + j >= 0 and i + j < len(ecg_diff):
                if ecg_diff[i] < ecg_diff[i + j]:
                    ecg_diff[i] = 0;
                    break;
    
    # return list of indices with identified peaks
    peaks = list()
    for i in range(len(ecg_diff)):
        if ecg_diff[i] != 0:
            peaks.append(i)
    
    # return peak location sequence
    return peaks
```

* **RR Intervals -** Plot the original (synthetic) ECG signal as well as the locations of the peaks to verify your method. Keep in mind to compensate for any delay you might have introduced by filtering. From the peak binary signal, estimate the $RR$ interval sequence $r_n$ and its average value as:

$$ \bar R = \frac{1}{N}\sum_{n=0}^N r_n, $$

where $N$ is the number of peaks detected. Provide an estimate of the average heart beat rate in beats per minute $[bpm]$.


```python
# estimate peaks
peaks = peak_detector(final_signal, 4)


# plot signals (superimpose detected peaks)


# compute rn sequence
intervals = list()
for i in range(len(peaks) - 1):
    intervals.append((peaks[i + 1] - peaks[i]))

# average RR interval
average_rr = np.mean(intervals) / s_frequency
print("Average RR interval: ", average_rr, "s")

# print average heart rate
print("Average heart rate: ", 60 / average_rr, "bpm")
```

```python
# data loading
with open('ecg_data/ecg_mitarr.json','r') as infile:
        data = json.load(infile)

# assign to variables
# using absolute value of original signal
input_signal = np.abs(data['voltage'])
frequency = data['fs']
input_time = np.linspace(0, len(input_signal) / frequency, len(input_signal))

# display signal
plt.plot(input_time, input_signal)
plt.title("Original Arrhythmia Signal")
plt.ylabel("Voltage (mV)")
plt.xlabel('Time (s)')

# estimate peaks, RR sequence, heart rate, and plot intermediate and final results
vtach_peaks = peak_detector(input_signal, 3.5)

# create list for all intervals
vt_intervals = list()

# calculate intervals as differences in indices of peaks
for i in range(len(vtach_peaks) - 1):
    vt_intervals.append((vtach_peaks[i + 1] - vtach_peaks[i]))

# we had many adjacent peaks which added undesirable intervals of small values
# so we used a thresholding to remove these to ensure an accurate calculation
vt_clean_intervals = list()
for i in range(len(vt_intervals)):
    if vt_intervals[i] > 7:
        vt_clean_intervals.append(vt_intervals[i])
    else:
        vt_clean_intervals.append(0)
        
# create list for instantaneous heart rate
inst_heart_rate = list()

# calculate instantaneous heart rate for time series
for i in range(len(vt_clean_intervals)):
    # for non-zero interval values, calculate the bpm
    if vt_clean_intervals[i] != 0:
        inst_heart_rate.append(60 / (float(vt_clean_intervals[i]) / frequency))



        
# plot instantaneous heart rate
plt.figure()
inst_hr_time = np.linspace(0, len(input_signal) / frequency, len(inst_heart_rate))
plt.plot(inst_hr_time, inst_heart_rate)
plt.title('Instantaneous Heart Rate')
plt.xlabel('Time (s)')
plt.ylabel('Heart Rate (beats per minute)')


# we seemed to have a very high spike for instantaneous heart rate
# which is almost definitely inaccurate. This may result from two 
# peaks that are very close together. This is why we decided to implement 
# the thresholding for the intervals, since short intervals result in very high
# instantaneous bpms, but too high of a threshold seemed to make our output worse
```

### Denoising
Now let us consider a more realistic scenario where we have a noisy ECG signal. There are different
sources of noise that can be present during an ECG acquisition. In addition to high-frequency thermal
noise, it is also common to observe the presence of low-frequency interference coming from breathing. 
When it comes to pathologies, different non-additive distortions might be present on the ECG signal 
that alter the shape of the QRS-complex itself but for the purpose of this task we just assume a healthy 
individual where signal distortion comes solely from the acquisition process. The procedure that we
will employ in order to estimate the locations of the QRS-complex is based on the Pan-Tompkins
algorithm [Pan-Tompkins algorithm](https://ieeexplore.ieee.org/stamp/stamp.jsp?tp=&arnumber=4122029).

* **Effect of noise** - Load the signal **ecg_mitnst.json** and plot the signal over time. The signal corresponds to a sample ECG from the [MIT noise stress dataset](https://physionet.org/physiobank/database/nstdb/)
that has been downloaded from [physionet.org](https://physionet.org/physiobank/database/).
Plot the signal and observe the presence of a strong low-frequency component.

```python
# load data from disk
with open('ecg_data/ecg_mitnst.json','r') as infile:
        data = json.load(infile)


# assign to variables
noisy_data = data['voltage']
samp_freq = data['fs']

# create time list for plotting
time = np.linspace(0, len(noisy_data) / samp_freq, len(noisy_data))


# plot signal with annotations
plt.plot(time, noisy_data)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
```

* **Pan-Tompkins Algorithm -** In order to deal with nose we will implement a simplified version of the Pan-Tompkins algorithm. For that purpose proceed as follows:

  1. _Bandpass filtering:_ Follow the steps described in the manual in order to perform a bandpass filtering operation as the concatenation of lowpass and highpass filtering operations. Display the obtained signals at the intermediate steps.
  
  2. _Differentiation:_ Use the discrete filter $d_n$ to approximate the derivative of the underlying signal:
  
    $$ d_n = \frac{1}{8}\big(\delta_{n-2} + 2\delta_{n-1} -2\delta_{n+1} - \delta_{n+2}\big) $$
  
    The filter is intended to localize the steepest region in the QRS-complex.
  
  3. _Signal squaring:_ Square the obtained signal after differentiation and plot the obtained waveform. On the squared signal use an averaging filter of length $L=20$. Why was squaring helpful in revealing the peaks of the QRS complex?
  
  3. _Integration:_ Integrate the resulting signal from the squaring operation with a box window of length $L=50$. Display the resulting signal.
  
  4. _Peak detection:_ Use a peak detector to reveal the locations of the peaks. Estimate the average $RR$ interval and plot the sequence of estimated peaks.
 

```python
# follow steps 1 to 5, plot intermediate results and sequences

# normalize noisy data
# calculate statistics
signal_mu = np.mean(noisy_data)
signal_std = np.std(noisy_data)

# turn every data point into a z-score
noisy_data = (noisy_data - signal_mu) / signal_std

# create box filter with length of the sample frequency
box_length = 360
box = signal.boxcar(box_length)


# create a low pass filter by convolving a box with itself
low_pass_filter = np.convolve(box, box, mode = "same")
# partially normalize the filter by dividing it by its total sum
low_pass_filter = low_pass_filter / np.sum(low_pass_filter)

# determine the low frequencies from the signal using the low pass filter
noisy_data_lows = np.convolve(low_pass_filter, noisy_data, mode = "same")
lows_time = np.linspace(0, len(noisy_data) / samp_freq, len(noisy_data))
plt.figure()
plt.plot(lows_time, noisy_data_lows)
plt.title('Signal After Using Low Pass Filter')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')

# find the high frequencies from the signal by subtracting the low frequencies
noisy_data_highs = noisy_data - noisy_data_lows
plt.figure()
plt.plot(lows_time, noisy_data_highs)
plt.title('Signal After Eliminating Low Frequencies')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')


# create bandpass filter with length 10
bandpass_box_length = 10
bandpass_filter = signal.boxcar(bandpass_box_length)

# find the signal after the use of the bandpass filter
noisy_data_bandpass = np.convolve(bandpass_filter, noisy_data_highs, mode="same")

# plot the signal after using the bandpass filter
plt.figure()
plt.plot(lows_time, noisy_data_bandpass)
plt.title('Signal After Eliminiating Both High and Low Frequencies')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')


# create deriviative filter according to instructions
derv_filter = list() 
derv_filter.append(float(1/8))
derv_filter.append(float(2/8))
derv_filter.append(float(0))
derv_filter.append(float(-2/8))
derv_filter.append(float(-1/8))

# calculate signal after using derivative filter
noisy_data_derv = np.convolve(derv_filter, noisy_data_bandpass, mode="same")
plt.figure()
plt.plot(lows_time, noisy_data_derv)
plt.title('Derivative of Signal')
plt.xlabel('Time (s)')
plt.ylabel('Change in Voltage (mV/s)')

# square deriviative of signal
noisy_data_derv = [i ** 2 for i in noisy_data_derv]
# plot results
plt.figure()
plt.plot(lows_time, noisy_data_derv)
plt.title('Square of Derivative of Signal')
plt.xlabel('Time (s)')
plt.ylabel('Change in Voltage (mV/s)')

# create averaging filter with length 20
average_filter_length = 20
average_filter = signal.boxcar(average_filter_length)

# calculatae averaged signal
noisy_data_derv = np.convolve(noisy_data_derv, average_filter, mode = "same")

plt.figure()
plt.plot(lows_time, noisy_data_derv)
plt.title('Averaged Square of Derivative of Signal')
plt.xlabel('Time (s)')
plt.ylabel('Change in Voltage (mV/s)')

# create integration filter of length 50
integration_length = 50
integral_filter = signal.boxcar(integration_length)

# calculate integrated signal
clean_data = np.convolve(noisy_data_derv, integral_filter, mode = "same")

plt.figure()
plt.plot(lows_time, clean_data)
plt.title('Final Processed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Change in Voltage (mV/s)')

# approximate peaks of filtered signal
clean_data_peaks = peak_detector(clean_data, 1.3)

# create list for original signal but only peaks
clean_data_only_peaks = list()

for i in range(len(clean_data_peaks)):
    clean_data_only_peaks.append(noisy_data[clean_data_peaks[i]])
    
plt.figure()
plt.plot(noisy_data)

# estimating average RR interval and HR
hr_intervals = list()

# find intervals between peaks
for i in range(len(clean_data_peaks) - 1):
    hr_intervals.append((clean_data_peaks[i + 1] - clean_data_peaks[i]))

# threshold intervals for reasons previously discussed
hr_clean_intervals = list()
for i in range(len(hr_intervals)):
    if hr_intervals[i] > 14:
        hr_clean_intervals.append(hr_intervals[i])
    else:
        hr_clean_intervals.append(0)

# retain only non-zero intervals
final_intervals = list()
for i in range(len(hr_clean_intervals)):
    if hr_clean_intervals[i] != 0:
        final_intervals.append(hr_clean_intervals[i])
        
# calculate average RR interval
average_interval = (np.mean(final_intervals)) / samp_freq

# report averate RR interval
print("Average RR Interval: ", average_interval)

        
inst_heart_rate = list()

# calculate instantaneous heart rate 
for i in range(len(hr_clean_intervals)):
    if hr_clean_intervals[i] != 0:
        inst_heart_rate.append(60 / (float(hr_clean_intervals[i]) / samp_freq))
        
# report average heart rate
print('Average heart rate (BPM):', (60 / average_interval))

final_peak_indices = list()

# find indices of only non-zero interval peaks
for i in range(len(hr_clean_intervals)):
    if hr_clean_intervals[i] != 0:
        final_peak_indices.append(clean_data_peaks[i])
        
# plot superimposition on original signal
plt.plot(clean_data_peaks, clean_data_only_peaks, 'ro')
plt.title('Original Signal With Detected Peaks')
plt.xlabel('Time (s)')
plt.ylabel('Change in Voltage (mV/s)')
```
